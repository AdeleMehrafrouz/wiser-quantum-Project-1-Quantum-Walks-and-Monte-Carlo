```python
#run_all.ipynb
# Quantum Galton Box Challenge

from circuits.galton_generator import simulate_distribution
from circuits.biased_distribution import generate_biased_galton_box, run_simulation, postprocess_counts
from metrics.distance_metrics_multi import total_variation_distance, kl_divergence
import matplotlib.pyplot as plt
import numpy as np
```

```python
layers = 4
gauss = simulate_distribution(layers)
biased = simulate_distribution(layers, biased=True, theta=2 * np.pi / 3)
```

```python
plt.bar(gauss.keys(), gauss.values(), alpha=0.6, label='Gaussian')
plt.bar(biased.keys(), biased.values(), alpha=0.6, label='Exponential (Biased)')
plt.legend()
plt.title("Galton Box Output Distributions")
plt.xlabel("Bin Index")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

```python
print("TVD:", total_variation_distance(gauss, biased))
print("KL Divergence:", kl_divergence(gauss, biased))
```
